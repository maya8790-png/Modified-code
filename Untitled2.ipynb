```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
import matplotlib.pyplot as plt
import numpy as np

# ----------------------------
# 1. تعريف نموذج Self-Attention U-Net 3D
# ----------------------------
class SelfAttentionUNet3D(nn.Module):
    def __init__(self, in_channels=1, out_channels=1):
        super(SelfAttentionUNet3D, self).__init__()
        # Encoder
        self.encoder = nn.Sequential(
            nn.Conv3d(in_channels, 64, kernel_size=3, padding=1),
            nn.BatchNorm3d(64),
            nn.ReLU(inplace=True),
            nn.Conv3d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm3d(128),
            nn.ReLU(inplace=True)
        )
        # Decoder
        self.decoder = nn.Sequential(
            nn.ConvTranspose3d(128, 64, kernel_size=3, padding=1),
            nn.BatchNorm3d(64),
            nn.ReLU(inplace=True),
            nn.ConvTranspose3d(64, out_channels, kernel_size=3, padding=1),
            nn.Sigmoid()  # المخرجات Grayscale بين 0 و1
        )

    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded

# ----------------------------
# 2. إعداد بيانات التدريب والاختبار
# ----------------------------
class OCTMRIDataset(Dataset):
    def __init__(self, ct_images, mri_images, transform=None):
        self.ct_images = ct_images    # قائمة من صور CT (مصفوفات numpy)
        self.mri_images = mri_images  # قائمة من صور MRI (الحقيقية)
        self.transform = transform

    def __len__(self):
        return len(self.ct_images)

    def __getitem__(self, idx):
        ct = self.ct_images[idx]
        mri = self.mri_images[idx]
        if self.transform:
            ct = self.transform(ct)
            mri = self.transform(mri)
        return ct, mri

# تحويل بيانات 3D: نحول مصفوفة numpy إلى Tensor ونضيف بُعد القناة
def to_tensor_3d(x):
    return torch.tensor(x, dtype=torch.float).unsqueeze(0)

# نستخدم التحويل السابق مباشرةً (بياناتنا مُولدة بالحجم 128x128x128)
transform = to_tensor_3d

# إنشاء بيانات وهمية (dummy data)
dummy_ct = [np.random.rand(128, 128, 128) for _ in range(10)]
dummy_mri = [np.random.rand(128, 128, 128) for _ in range(10)]
dataset = OCTMRIDataset(dummy_ct, dummy_mri, transform=transform)
train_loader = DataLoader(dataset, batch_size=1, shuffle=True)

# ----------------------------
# 3. تدريب النموذج مع إمكانية استخدام Gradient Loss
# ----------------------------
def train_model(model, use_gradient_loss=False, num_epochs=5):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    optimizer = optim.Adam(model.parameters(), lr=0.0001, betas=(0.5, 0.999))
    criterion = nn.MSELoss()

    # دالة حساب خسارة التدرج المعدلة لحساب التدرج على الأبعاد المكانية فقط
    def gradient_loss(output, target):
        grads_output = torch.gradient(output, dim=(2,3,4), edge_order=1)
        grads_target = torch.gradient(target, dim=(2,3,4), edge_order=1)
        grad_output_mag = sum(torch.abs(g) for g in grads_output)
        grad_target_mag = sum(torch.abs(g) for g in grads_target)
        return torch.mean(torch.abs(grad_output_mag - grad_target_mag))

    for epoch in range(num_epochs):
        model.train()
        epoch_loss = 0
        for ct, mri in train_loader:
            ct, mri = ct.to(device), mri.to(device)
            optimizer.zero_grad()
            outputs = model(ct)
            loss = criterion(outputs, mri)
            if use_gradient_loss:
                loss += gradient_loss(outputs, mri)
            loss.backward()
            optimizer.step()
            epoch_loss += loss.item()
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss/len(train_loader):.4f}")
    return model

# تدريب نموذج U-Net 3D العادي
model_unet = SelfAttentionUNet3D()
print("تدريب U-Net 3D العادي:")
trained_unet = train_model(model_unet, use_gradient_loss=False, num_epochs=5)

# تدريب نموذج Self-Attention U-Net 3D مع Gradient Loss
model_sa_unet = SelfAttentionUNet3D()
print("تدريب Self-Attention U-Net 3D مع Gradient Loss:")
trained_sa_unet = train_model(model_sa_unet, use_gradient_loss=True, num_epochs=5)

# ----------------------------
# 4. اختبار النموذج وعرض النتائج
# ----------------------------
def visualize_results(ct, generated, real):
    # لضمان عرض 3D بطريقة مفهومة، نأخذ شريحة وسطية من حجم الصورة
    slice_idx = ct.shape[1] // 2
    plt.figure(figsize=(15,5))
    plt.subplot(1,3,1)
    plt.imshow(ct.squeeze()[slice_idx], cmap='gray')
    plt.title("CT Input")
    plt.axis('off')
    plt.subplot(1,3,2)
    plt.imshow(generated.squeeze()[slice_idx], cmap='gray')
    plt.title("Generated MRI")
    plt.axis('off')
    plt.subplot(1,3,3)
    plt.imshow(real.squeeze()[slice_idx], cmap='gray')
    plt.title("Real MRI")
    plt.axis('off')
    plt.show()

def test_model(model):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()
    ct_sample, mri_sample = dataset[0]
    ct_sample = ct_sample.unsqueeze(0).to(device)
    with torch.no_grad():
        generated_mri = model(ct_sample)
    ct_np = ct_sample.cpu().numpy()[0]
    generated_np = generated_mri.cpu().numpy()[0]
    mri_np = mri_sample.cpu().numpy() if isinstance(mri_sample, torch.Tensor) else mri_sample
    visualize_results(ct_np, generated_np, mri_np)

print("عرض النتائج:")
test_model(trained_sa_unet)
```
